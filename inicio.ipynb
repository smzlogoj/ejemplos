```r
df_raw <- read.csv('Data_Cortex_Nuclear.csv')
```

```r
head(df_raw)
```

```r
summary(df_raw)
```

```r
library(naniar)
vis_miss(df_raw)
```

```r
#Contar NAs por filas
row_na <- apply(is.na(df_raw), 1, sum)
#Contar NAs por columnas
col_na <- apply(is.na(df_raw), 2, sum)

row_na
col_na
```

```r
col_na[col_na > 10]
```

```r
df <- df_raw[, !names(df_raw) %in% names(col_na[col_na > 10])]
vis_miss(df)
df <- df[complete.cases(df),]
vis_miss(df)
```

```r
head(df)
```

```r
df$class <- gsub('-', '', df$class)
df$Behavior <- gsub('/', '', df$Behavior)
```

```r
head(df)
```

```r
Mouse <- c()
Sample <- c()
for (i in strsplit(df$MouseID, '_')){
    Mouse <- c(Mouse, i[1])
    Sample <- c(Sample, i[2])
}
df$Mouse <- Mouse
df$Sample <- Sample
```

```r
row.names(df) <- df$MouseID
df$MouseID <- NULL
head(df)
```

```r
proteinas <- names(df[, 1:69])
proteinas
```

```r
df <- df[, c('Mouse', 'Sample', 'Genotype', 'Treatment', 'Behavior', 'class', proteinas)]
head(df)
```

```r
df$Mouse <- as.factor(df$Mouse)
df$Sample <- as.factor(df$Sample)
df$Genotype <- as.factor(df$Genotype)
df$Treatment <- as.factor(df$Treatment)
df$Behavior <- as.factor(df$Behavior)
df$class <- as.factor(df$class)

df
```
